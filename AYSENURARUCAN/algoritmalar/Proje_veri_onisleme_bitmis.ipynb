```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("C:\\Users\\Dell\\OneDrive\\Masaüstü\\mlodev\\tablet.xlsx")
```

```python
df = pd.DataFrame()
```

```python
data.head()
```

```python
df['url'] = data['url']
df['Isletim_Sistemi'] = data['İşletim Sistemi']
df['Islemci_Tipi'] = data['İşlemci Tipi']
df['Cozunurluk'] = data['Çözünürlük']
df['Bellek_Kapasitesi'] = data['Bellek Kapasitesi']
df['Fiyat'] = data['price']
```

```python
df.head()
```

```python
# Fiyat sütunundaki değerleri düzenle
for i in range(len(df['Fiyat'])):
    # Virgül ve ondalık haneleri kaldır
    cleaned_price = ''.join(c for c in str(df['Fiyat'][i]) if c.isdigit() or c == '.')

    # Nokta işaretlerini kaldır
    cleaned_price = cleaned_price.replace('.', '')

    # Ondalık sayıya dönüştür
    df.at[i, 'Fiyat'] = float(cleaned_price) if cleaned_price else None
```

```python
df
```

```python
df["Bellek_Kapasitesi"] = df.Bellek_Kapasitesi.str.replace('GB', '')
```

```python
df
```

```python
df["Fiyat"] = df.Fiyat.astype(float)
```
