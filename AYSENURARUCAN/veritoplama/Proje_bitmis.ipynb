```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url_list = []
prices_list = []
propTitles = []
propValues = []

for i in range(1, 40):
    url = "https://www.trendyol.com/tablet-x-c103665?page=" + str(i)
    r = requests.get(url)
    source = BeautifulSoup(r.content, "lxml")

    urls = source.find_all("div", attrs={"class": "p-card-chldrn-cntnr card-border"})
    for url in urls:
        url_tablet = "https://www.trendyol.com/" + url.a.get("href")
        url_list.append(url_tablet)

        r_tablet = requests.get(url_tablet)
        source_tablet = BeautifulSoup(r_tablet.content, "lxml")

        properties = source_tablet.find_all("li", attrs={"class": "detail-attr-item"})
        for prop in properties:
            prop_title = prop.find("span").text.strip() 
            prop_value = prop.find_all("span")[1].text.strip() 
            propTitles.append(prop_title)
            propValues.append(prop_value)
        
    prices = source.find_all("div", attrs={"class": "prc-box-dscntd"})
    for price in prices:
        prices_list.append(price.text)

print(str(len(url_list)) + "Adet link bulundu")
print(str(len(prices_list)) + "Adet fiyat bulundu")
print(str(len(propTitles)) + "Adet başlık bulundu")
print(str(len(propValues)) + "Adet özellik bulundu")
```

```python
#Url ve fiyatları bir data frame yazma
df_urls = pd.DataFrame()
df_urls["urls"] = url_list
df_urls["prices"] = prices_list
```

```python
#Bulunan veri sayısı
tablet = len(url_list)
```

```python
#Bulunan özellik başlıklarının benzersizlerini bulma
columns = np.array(propTitles)
columns = np.unique(columns)
```

```python
#Başlıkları kullanarak url ve fiyat ile birlikte yeni bir data frame oluşturma
df = pd.DataFrame(columns=columns)
df["url"] = url_list
df["price"] = prices_list
```

```python
#Oluşturulan data frame gösterme
df.head()
```

```python
for i in range(0, tablet):
    url = df['url'].loc[i]  # Her döngü adımında farklı bir URL alınmalı
    r = requests.get(url)
    source = BeautifulSoup(r.content, "lxml")

    properties = source.find_all("li", attrs={"class": "detail-attr-item"})
    for prop in properties:
        prop_title = prop.find("span").text.strip() 
        prop_value = prop.find_all("span")[1].text.strip() 
        print(prop_title + ":" + prop_value)
        df[prop_title].loc[i] = prop_value  # Buradaki boşlukları kontrol edin

# Bu satırın sonunda boş bir satır bırakın
```
